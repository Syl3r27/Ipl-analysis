```python
df = pd.read_csv("matches.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(5,10))
plt.style.use('dark_background')
sns.countplot(y='city', data=df, order=df['city'].value_counts().index, palette='tab10')
plt.title('Number of Matches per City')
plt.xlabel('Number of Matches')
plt.ylabel('City')
plt.show()


```

```python
plt.figure(figsize = (10,10))
plt.style.use('dark_background')
top_players = df['player_of_match'].value_counts().head(10)
sns.barplot(x=top_players.index, y=top_players.values,palette='tab10')
plt.title('Top 10 Players of the Match')
plt.xlabel('Player')
plt.ylabel('Number of Matches')
plt.show()
```

```python
toss_decision_counts = df['toss_decision'].value_counts()
plt.style.use('ggplot')
plt.figure(figsize=(5,5))
plt.pie(toss_decision_counts, labels=toss_decision_counts.index, autopct='%1.1f%%', colors=['#FFC154', '#EC6B56'])
plt.title('Toss Decision Distribution')
plt.show()
```
